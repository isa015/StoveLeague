## 3.1 금리형 상품 개괄

##3.1.1 채권 이외의 금리부 상품들   


* (정기) 예금   
> 발행 주체 : 은행   
이자 부리 방식 : 단리   
발행 방식 : 액면 발행   
거래 유동성 : 거래 불가, 중도 해지만 가능    
리스크 : 마켓 리스크 없음(시가평가 대상이 아님), 신용 리스크는 있음   
기타 : 증권사 발행어음

* 양도성예금증서 (CD, Certificate of Deposit)
> 발행 주체 : 은행   
이자 부리 방식 : 단리    
발행 방식 : 액면 발행   
거래 유동성 : 거래 가능    
리스크 : 마켓 리스크 있음(시가평가 대상), 신용 리스크 있음   

* 은행인수어음 (BA, Banker's Acceptances)
> 발행 주체 : 은행 (비은행의 무역거래 어음에 대한 보증)   
이자 부리 방식 : 선취 할인    
발행 방식 : 할인 발행   
거래 유동성 : 거래 가능    
리스크 : 마켓 리스크 있음(시가평가 대상), 신용 리스크 있음   

* 기업어음 (CP, Commercial Paper)
> 발행 주체 : 기업   
이자 부리 방식 : 선취 할인 : 할인율 vs. 실효수익율  
발행 방식 : 할인 발행   
거래 유동성 : 거래 가능    
리스크 : 마켓 리스크 있음(시가평가 대상), 신용 리스크 있음   
기타 : 기업어음 발행자에 대한 신용등급 체계 (채권과 별도), 1년 이내 만기가 일반적(!)


* 환매조건부채권 (Repurchase Agreement, RP 혹은 Repo) 매도 : 대고객 RP
> 발행 주체 : 금융기관(대체로 증권사) (담보부 금전 대차)  
이자 부리 방식 : 단리    
발행 방식 : 액면 발행   
거래 유동성 : 거래 불가, 중도 해지만 가능    
리스크 : 마켓 리스크 없음(시가평가 비대상), 신용 리스크 있음(발행자/담보자산)   
기타 :   
>> 1) 발행자 입장에서, RP 매도는 담보부 금전 차입   
   2) 대고객 RP vs. 기관간 RP   
   3) Repo 는 통상 자금차입(Repo 매도)을, Reverse Repo 는 자금대여(Repo 매수)를 의미하나, 반대의 의미로도 쓰이는 경우가 자주 있음   
   4) 중앙은행의 유동성 조절 수단으로서도, 자본시장에서의 레버리지 수단으로서도 매우 중요   


* 금전신탁
> 운용 주체 : 수익자(일임계약에 의해 은행/증권사가 대행)   
운용 특징 : 확정 수익이 아닌 성과 배당

* 펀드 / ETF
> 운용 주체 : 운용사   
운용 특징 : 확정 수익이 아닌 성과 배당

##3.1.2 채권(債券, Fixed income Securities)   (1)

* 채권의 발행 정보 (글로벌채권 기준)   
> * Issuer’s name : (예) Bundesrepublik Deutschland   
> * Issuer’s type : (예) the oil sector   
> * Market in which the bond is issued : 개별 국가, eurodollar 시장
> * Issuer’s domicile : 발행사의 소속 국가
> * Bond’s currency denomination : 발행 통화   
> * Method used for the calculation of the bond price/yield : 한국은 관행적 복할인 방식  
> * Type of guarantee : 담보 / 국가   
> * Maturity date : 만기일   
> * Coupon type : 이표 금리 결정 방식 : fixed / floating / multicoupon (Step Up 조항 등)   
> * Coupon rate  : 발행시 이표 금리   
> * Coupon frequency : 이자 지급 주기 : 미국/한국 국채 반기   
> * Interest accrual date : 이자 기산일   
> * Settlement date : 거래후 결제일 : 한국은 T+1일 글로벌채권은 T+2일 이상   
> * Spread at issuance : 동일만기 국채(UST) 금리 대비 발행 스프레드   
> * Identifying code : 개별 채권 식별 코드 : ISIN (International Securities Identification Number) and the CUSIP (Committee on Uniform Securities Identification Procedures)   
> * Total issued amount / Outstanding amount : 발행금액 / 발행잔액   
> * Rating : Moody's / S&P / Fitch 등 : 국가 및 기업(아람코)
