* 채권 금리 수준에 대한 타이밍 베팅   

 * 금리 하락 예상시 / 금리 상승 예상시

   * 듀레이션 확대, 만기가 길고, Coupon 금리가 낮은 종목 중심으로 편입
   
   * 듀레이션 축소, 롤오버 (RollOver)   
   
   *  모멘텀(Momentum)과 리버설(Reversal)   
    
     * 모멘텀 : 채권 시장이 강해지면 포지션을 늘리고, 약해지면 줄인다   
     * 리버설 : 일정 범위를 넘어서면 시장 흐름에 반대로 포지션을 잡는다

* 수익률 곡선의 특정한 변화에 대한 베팅   

  * **Bullet Portfolio** : 특정 만기(중기)에 집중하여 투자

  * **Barbell Portfolio** : 수익률 곡선의 양쪽 날개(단기와 장기)로 분할하여 구성

     * 같은 듀레이션 하에서는, Barbell이 Bullet 보다 Convexity가 큼

  * **Ladder Portfolio** : 만기별로 고르게 배분하여 구성

  * Butterfly 전략

    * Barbell (Wing, Long)과 Bullet (Body, Short)의 결합   
    * Portfolio 는 순현금 0 (cash-neutral) 이거나 $듀레이션 0 --> Convexity Only   
    * (예) 2년 및 30년 Long, 10년 Short (2개)으로 구성   
    * 수익률 곡선이 수평으로 움직이면 (2년,10년, 30년 금리가 같은 폭 변동) --> 금리 방향과 무관하게 + 수익, 변동이 클수록 큰 수익

## 4.1.2 시장 구조에 기반한 트레이딩`

* 상대가치(Relative Value) 혹은 스프레드(Spread) 전략   

  * Long-Shrot 전략 : 유사성이 있는 자산간에 시행   

  * 두 금리 자산 간의 금리차이(Spread) 움직임이 평균회귀(Mean-Revision) 패턴을 보인다는 점에 주로 착안

     * 포지션을 장기간 유지하려 하는 경우에는 두 자산 간의 수익률 곡선도 비교   
     
     * 대체로, 금리가 높거나 향후 낮아질 것으로 예상되는 금리를 Long, 반대를 Short   

     * 다른 자산과는 달리, 채권의 Long-Short 전략은 Positive Carry가 가능한 경우가 많음   

  * 현금 소요가 없거나 작고, 측정상의 듀레이션 노출이 없으면서도 양(Positive)의 보유수익(Carry)

  * Carry 의 절대적인 크기가 작아도 자기자본 투자 금액 대비 ROE가 높음    

  * Repo 매도를 활용하여 레버리지 포지션을 만들어낼 수 있음   
    * 네이키드(Naked) 레버리지 포지션도 가능   
      : 삼성카드(AA+) 2년 2.70% / Repo 2.50% -> 2.70 + (2.70 - 2.50) = 2.90 %    

  * 채권 현물-선물 베이시스 트레이딩, 버터플라이 트레이딩, IRS 스프레드 트레이딩, 채권-IRS 트레이딩   

* Risk and Return in Fixed Income Arbitrage: Nickels in Front of a Steamroller?   
  
    * 스프레드 축소에 베팅했는데 반대 상황이 발생하면 ?  시간과의 싸움  
    * Short Gamma 속성 : 시장 변동성이 커질 때 손실이 커지는 경향
