##3.3 채권 파생상품

* 국채 선물   

> : 한국 : 쿠폰 5% 의 가상적인 3년(10년) 만기 국채에 대한 선물 : 선물 만기일에도 채권 현물 인수도 없음 : 호주 방식   
> * 3년 선물, 10년 선물   

> : 미국 : 현물 국채에 대한 선물 : 선물 만기일에 채권 현물 인수도(Physical delivery) 있음   

> * Basis : (한국) (가상 채권의) Spot Clean Price - 선물가   
(미국) Spot Clean Price - 선물가*Conversion factor   

> * Basis Trading     

> : 일일 정산, 거래소에서 거래되는 장내 파생상품 (거래상대방 리스크 없음)


* 이자율 스왑 (Interest Rate Swap : IRS)   

> * 고정금리로 계산된 현금흐름(Fixed leg)과 변동금리로 계산된 현금흐름들(Floating leg)을 일정한 시점에 교환하기로하는 계약   
> * 변동금리 : CD (KRW), KOFR(KRW), LIBOR (USD), SOFR (USD)   
> * 플레인 바닐라 Plain Vanilla 형태의 경우, 중앙청산결제소(CCP : Central Counterparty)에서 거래 --> 장내 파생상품, 일일 정산   
> * 초기 및 만기에 원금 교환이 없음. 거래 기간 중 이자만 교환   
> * **경제적 실질 : KRW 고정금리 채권와 KRW FRN간 Trading**   
>> * FRN의 경제적 실질 : [고정금리 채권 + IRS Pay] 조합  

> * 호칭 : 고정금리 파트 기준으로 호칭 : Pay (fixed), Receive (fixed)   
>> * Pay (fixed) : 원화 고정금리 채권 Short / 원화 FRN --> 일반적으로 금리가 오를 때 수익   


* 이자율 스왑의 거래 형태들   
 * Outright 거래   
 * Spread 거래 : 서로 다른 만기의 두 IRS를 Pay/Receive를 반대로 하여 동시에 거래   
 >* 호칭 : 장기물 기준으로 호칭   
 >* 두 거래의 DV01 의 합이 0 이 되도록 거래 : Delta neutral  
: (예) 3년 IRS 액면 100억의 DV01 = 292 만원, 5년 IRS 액면 100억의 DV01 = 477 만원 --> 5년물 액면 100억 / 3년물 액면 160억   
>* [3/5년 스프레드 100 억 Receive]=[3 by 5 100억 Receive]=[3's5's 100억 Receive] = 5년 IRS 액면 100억 Receive & 3년 IRS 액면 160억 Pay   

 * Butterfly 거래   
 >* 두 Spread 거래의 합(방향은 반대)   
 >* 2-3-5 butterfly long : [2 by 3] Pay & [3 by 5] Receive

##3.4 채권 FX 파생상품

* 통화 스왑 (CCS : Cross Currency Interest Rate Swap) : 서로 다른 통화의 현금흐름 교환     
 * 호칭 : 고정금리 기준
 * CCS Pay : KRW 고정 금리를 지급, USD 변동금리 수취  
 : **경제적 실질 : KRW 고정 금리 채권 매도(Short) vs. USD FRN(LIBOR/SOFR) 매수(Long)**   
 : 원화 조달 & USD 운용   
 : 외화조달시 환헤지를 위한 Libility Swap
 * CCS Receive : KRW 고정 금리를 수취, USD 변동금리 지급  
 : **경제적 실질 : KRW 고정 금리 채권 매수(Long) vs. USD FRN 매도(Short)**   
 : USD 조달 & 원화 운용   
 : 국내 시장 달러화 부족 --> CCS Receive 수요 증가 --> CCS 금리 하락   
 : 해외채권 투자시 환헤지를 위한 Asset Swap  
 * Basis Swap : 원화 FRN 채권 vs. USD FRN


* FX 스왑 : 처음과 끝에만 현금흐름 교환
 * 호칭 : 달러화 Spot 기준   
 * Buy&Sell : USD를 Spot 환율 기준으로 사고, 계약 종료시점에 선도(Forward) 환율 기준으로 USD 매도하기로 하는 계약   
 : 경제적 실질 : USD 단기 복리채 매도(Short), KRW 단기 복리채 매수(Long)     
 : USD 조달 & KRW 운용   
 : 해외채권 투자시 환헤지를 위한 Asset Swap   
  * Sell&Buy : USD를 Spot 환율 기준으로 팔고 종료 시점에 Forward 환율 기준으로 USD를 매수하기로 하는 계약   
 : 경제적 실질 : USD 단기 복리채 매수(Long), KRW 단기 복리채 매수(Short)
 : KRW 조달 & USD 운용   
 : 외화조달시 환헤지를 위한 Libility Swap   
 * Swap Point (혹은 Swap Marging) : Forward 환율 - Spot 환율   
 : 거래 시작일 (Effective date) : USD notional vs. KRW (Spot 환율)   
 만기일 (Maturity date) : USD notional vs. KRW (Spot 환율 + Swap Point)   
 : 호가 방식 : 표기된 값에서 100 을 나누어 준 값   
 : 원화 금리가 더 높을 때는 Plus Swap Point, 낮을 때는 Minus   
 * 기본적으로, KRW/USD 통화간 금리차이에 의해 발생   
 : 원화 금리가 더 높을 때는 Plus Swap Point, 낮을 때는 Minus   
 : (예) Buy&Sell (달러 조달/원화 운용) : USD 금리 > KRW 금리 --> 높은 금리의 통화를 보유하는 댓가로서, 만기에 돌려받는 KRW 금액이 초기 금액보다 작은게 정상   
 * Covered Interest Rate Parity   
 : Swap Point가 재정거래(Arbitrage Trading)를 통해 KRW/USD 통화간 금리차이 만큼 정확하게 반영   
 : 국내에서는, Sell&Buy 거래를 통한 초과수익 기회가 지속   
 : Forward Sell 초과 수요 : 국내 수출기업들(조선 등)의 속성   
 : 외국계 은행 국내 지점들의 단기 외화차입 수요를 증가시키는 원인으로 작용
