##3.2.3 수익률 곡선의 모양에 대한 이론들

* 기대 가설 (Expectations Theory)   

   : 투자자들은 현재의 단기 금리와 미래 금리 변동을 고려하여 장기 금리를 결정. 만기가 긴 채권의 수익률은 향후 단기 금리가 상승할 것으로 예상되면 높아지고, 하락할 것으로 예상되면 낮아진다는 식의 설명   

   : (예) 가까운 시일 내 기준금리의 큰 폭 인하가 예상되면 장기 금리가 먼저 크게 낮아짐    

* 유동성 프리미엄 가설 (Liquidity Premium Theory)   

  : 기대 가설을 보완하는 개념. 장기 채권은 단기 채권보다 유동성이 낮아 투자자들이 추가적인 위험 프리미엄을 요구한다는 설명. 장기 채권의 수익률은 단기 채권의 기대 수익률뿐만 아니라 유동성 프리미엄을 포함하여 결정   

  : (예) 장기 국채 발행이 급증할 가능성이 높아지면 장기 금리가 크게 상승함   

* 분할 시장 가설 (Market Segmentation Theory)   
  : 투자자들이 특정 만기 구간에 대한 선호가 있어, 개별적인 수요와 공급이 수익률을 결정한다고 설명합니다. 단기, 중기, 장기 채권 시장이 각각 독립적으로 움직이며, 이로 인해 수익률 곡선의 모양이 결정      

  : (예) 국고채 : 초단기는 MMF, 단기는 외국인 수요 반영, 초장기는 보험사 수요 반영

##3.3.3 듀레이션(Duration)과 컨벡서티(Convexity)   



* 듀레이션   

> * Macaulay Duration (메컬레이 듀레이션)    


>>> : 채권의 개별 현금흐름들의 가중 평균 잔존 만기 : 년(Years)   
: 가중치는, 채권 가격 (=개별 현금흐름들 현재 가치의 총합)에서 개별 현금흐름의 현재 가치가 차지하는 비중   
: 현금 흐름이 1개인 경우는 만기와 동일    
: (수정) 듀레이션에 비해, 상대적으로 개념이 직관적이고, 컴퓨터가 없는 상황에서는 계산이 용이함. 수정 듀레이션에 대한 Proxy 성격   
: ALM (Asset Libility Management) 측면에서 부분적인 활용성이 있음    

> * 메컬레이 듀레이션 공식

>>> image.png

> * (수정) 듀레이션 (Modified Duration)    
>> : 채권 가격에 대한 1차 미분값(differential) : YTM 1 % point 바뀔때 가격 변화율 : (예) YTM 이 1 % point 오르면 채권 가격이 7% 하락함   
: 옵션의 델타(Delta)에 준하는 개념   
: 메컬레이 듀레이션과 연관성이 있으나 수정 듀레이션이 보다 대표적  
: 실무적으로는 차분(difference)을 주로 활용 : Effective 듀레이션

> * 수정 듀레이션 수식

* 컨벡서티 (Convexity)   

>> : 채권 가격에 대한 2차 미분값(differential) : 실무적으로는 차분   
: 옵션의 감마(Gamma)에 준하는 개념 : 클수록 좋다   
: 만기가 길어질 수록 값이 커짐 : 장기 채권 매수 = Convexity Buying   
: 금리가 높아질수록 값이 커짐 : Long Gamma

* US MBS (Mortgage Backed Security) 의 특수성   
> * 음 (-)의 Convexity : 장기 채권 시장 금리가 높아질수록 듀레이션이 커짐 (Short Gamma)   
: MBS 매수 : Convexity Selling  
: MBS 에 내재되어 있는 조기 상환 옵션의 가치가 시장 금리에 연동
: Short Gamma : 부정적인 금리 변동이 발생하면 delta가 더 커짐 --> MBS 금리가 좋은 이유
: 본래적 의미의 YTM, Duration, Convexity 는 Coupon 및 Maturity 가 고정인 채권에만 적용되나 유추하여 적용할 수 있습니다.   
: 장기 채권금리 상승 -->  MBS 의 조기 상환 가능성 감소 --> MBS 포트폴리오의 듀레이션 상승 --> 포트폴리오 리스크 관리를 위한 채권 매도 --> 장기 채권금리 상승
