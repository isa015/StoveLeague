##3.2.3 채권의 가격 특성 지표   



* 듀레이션   

> * Macaulay Duration (메컬레이 듀레이션)    


>>> : 채권의 개별 현금흐름들의 평균 잔존 만기 : 년(Years), 가중치가 부여된   
: 가중치는, 채권 가격 (=개별 현금흐름들 현재 가치의 총합)에서 개별 현금흐름의 현재 가치가 차지하는 비중   
: 현금 흐름이 1개인 경우는 만기와 동일    
: (수정) 듀레이션에 비해, 상대적으로 개념이 직관적이고, 컴퓨터가 없는 상황에서는 계산이 용이함. 수정 듀레이션에 대한 Proxy 성격   
: ALM (Asset Libility Management) 측면에서 부분적인 활용성이 있음    

> * 메컬레이 듀레이션 공식

>>> image.png

> * (수정) 듀레이션 (Modified Duration)    
>> : 채권 가격에 대한 1차 미분값(differential) : YTM 1 % point 바뀔때 가격 변화율 : (예) YTM 이 1 % point 오르면 채권 가격이 7% 하락함   
: 옵션의 델타(Delta)에 준하는 개념   
: 메컬레이 듀레이션과 연관성이 있으나 수정 듀레이션이 보다 대표적  
: 실무적으로는 차분(difference)을 주로 활용 : Effective 듀레이션

> * 수정 듀레이션 수식   

>>image.png


* 메컬레이 듀레이션과 수정 듀레이션   
> * 메컬레이 듀레이션 >= 수정 듀레이션   
> * 잔존 만기가 길어질수록 듀레이션이 증가   
> * 잔존 만기가 길어질수록 늘어나는 듀레이션 크기는 체감 : 10년 만기 vs. 30년 만기    

> Duration_Maturity.png





* 컨벡서티 (Convexity)   

>> : 채권 가격에 대한 2차 미분값(differential) : 실무적으로는 차분
: 옵션의 감마(Gamma)에 준하는 개념 : 클수록 좋다   
: 만기가 길어질 수록 값이 커진다   
: 금리가 높아질수록 값이 커진다 : Long Gamma   
: 장기 채권 매수 : Convexity Buying   

> Convexity_Maturity.png




Duration_YTM.png




* US MBS 의 특수성   
> * 음 (-)의 Convexity : 장기 채권 시장 금리가 높아질수록 듀레이션이 커짐 (Short Gamma)   
: MBS 에 내재되어 있는 조기 상환 옵션의 가치가 시장 금리에 연동
: Short Gamma : 부정적인 금리 변동이 발생하면 delta가 더 커짐 : MBS의 금리가 좋은 이유
: 본래적 의미의 YTM, Duration, Convexity 는 Coupon 및 Maturity 가 고정인 채권에만 적용되나 유추하여 적용   
: 장기 채권금리 상승 -->  MBS 의 조기 상환 가능성 감소 --> MBS 포트폴리오의 듀레이션 상승 --> 포트폴리오 리스크 관리를 위한 채권 매도 --> 장기 채권금리 상승

: 시간 경과를 고려
