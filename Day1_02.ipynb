def larger(num1, num2)
```   

 > 설명 : 함수는 입력된 내용을 받아 입력된 내용을 처리하고 출력을 반환. 함수를 사용하면 코드의 복잡성을 줄일 수 있음.   
메서드는 특정한 자료형에만 쓸 수 있는 함수를 지칭함   


* 변수   

 > 예시 :
```
age = 25
```

 > 설명 : 저장된 값을 컴퓨터가 사용할 수 있도록 부여하는 이름. 할당문(=기호)을 사용해 변수에 값을 할당할 수 있음   


* 조건   
>예시 :
```
if age < 20:   
    print("투표 불가")
else:
    print("투표 가능")
```   
설명 : 조건문을 사용하면 코드에 분기를 만들 수 있음. 파이썬에서 조건문 관련한 세 가지 키워드는 if, elif, else   


* 문자열   
> 예시 :
```
name = 'Dan'
```   
설명 : 텍스트를 저장하고 조직하는데 사용되는 자료형.    
문자열의 각 값에는 인덱스가 부여되며 인덱싱은 1이 아닌 0 부터 시작.    
문자열을 수정할 때 사용하는 별도의 메서드가 있음.   


* 리스트    

 > 예시 :
```
list = ['Leo', 'Dan']
```   
 설명 : 여러 개의 값에 순서를 부여해 하나의 값으로 저장(숫자 또는 문자열 여러 개를 한 번에 저장)할 수 있는 자료형.   
 리스트의 각 값에는 인덱스가 부여되며 인덱싱은 1이 아닌 0 부터 시작. 리스트를 수정할 때 사용하는 별도의 메서드가 있음.

## 2.4 파이썬 코드를 읽기 위해 알아야 할 것들(2)   



### 2.4.1 파이썬 코드 기능들



* 루프   

> 예시 :   

>```
for loop:   
  for country in countries:   
  print(country)
```
>```
while loop:   
  index = 0   
  while index < 4:   
    print(index)   
    index = index+1
```
   
> 설명 : 루프를 사용하면 동일한 코드를 필요한 횟수만큼 실행할 수 있음.    
반복 횟수를 알고 있는 경우는 for 루프, 모르는 경우는 while 루프를 사용   


* 들여쓰기   

> 예시 :    
> ```
for country in countries:   
  print(country)
```


> 설명 : 파이썬 인터프리터에게 해당 코드가 다른 코드 본문의 일부로 속해 있는 경우(print 문이 for 루프 내에 있는 경우)를 알려줌   


* 딕셔너리   

> 예시 :
> ```
points = {'a':1, 'b':3}
```   
설명 : 딕셔너리를 사용하면 키를 값과 연결할 수 있음. 키 'a'는 값 1과 연결   


* 파일

```
file = open('chores.csv')   
first_line = file.readline()
```   
설명: 파일에는 데이터가 포함. 이 파일을 열고 데이터를 읽어들임   


 * 모듈   
> 예시 :
```
